简单上传流程：

1.安装打包和上传工具

```bash
pip install setuptools wheel twine
```

2.打包项目到/dist文件下生成```.tar.gz```和```.whl```文件

```bash
python setup.py sdist bdist_wheel
```

3.注册或登录PyPI账号

4.上传到PyPI
```bash
twine upload dist/*
```

5.现在在比较新的项目中setup.py的一个替代方案是写一个```pyproject.toml```
    ```
toml
    [build-system]
    requires = ["setuptools", "wheel"]
    build-backend = "setuptools.build_meta"

    [project]
    name = "muon-optimizer"
    version = "0.1.0"
    authors = [{name = "Keller Jordan", email = "kjordan4077@gmail.com"}]
    description = "A powerful optimizer"
    readme = "README.md"
    dependencies = ["torch", "numpy"]

    ```

6.此时的打包指令会变成:
```bash
python -m build #--module
```
